```python
import pandas as pd
import seaborn as sns

# Load Iris dataset from seaborn's built-in datasets
iris = sns.load_dataset("iris")  # No need to download manually

# Show basic structure
print("ðŸ“Š Dataset Sample:")
display(iris.head())

print("\nðŸ“‹ Dataset Info:")
print(iris.info())

print("\nðŸ§® Missing Values:")
print(iris.isnull().sum())

print("\nðŸ“ˆ Summary Statistics:")
display(iris.describe())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', palette='Set2')
plt.title("Sepal Length vs Width by Species")
plt.grid(True)
plt.show()
```

```python
iris.hist(figsize=(10, 8), bins=15, color='green', edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris, palette="pastel")
plt.title("Box Plots for All Features")
plt.grid(True)
plt.show()
```
